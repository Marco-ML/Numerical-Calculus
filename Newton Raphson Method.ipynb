```python
import numpy as np
import pandas as pd

def newton_raphson(x0, stop):
    global c
    
    x = [x0]
    f_v = [x0]
    error = [x0]
    c = 0
    i = x0
    iterations = []
    stop_criteria = stop
    
    while i > stop_criteria:
        iterations.append(c)
        f =  x[c]**4 + 1.904*1e15*x[c] - 6.04*1e20
        f_v.append(f)
        df = 4*x[c]**3 + 1.904*1e15
        x.append(x[c] - (f)/(df))
        c = c + 1
        i = np.abs(x[c] - x[c - 1])
        error.append(i)
    return (x, error)

def data_frame():
    x, error = newton_raphson(2.377*1e5, 1e-4)

    elements_df = [x, error]

    elements_np = np.array(elements_df).T

    elements_list = elements_np.tolist()

    df = pd.DataFrame(elements_list, columns = ['x', 'Error'], index = range(1, c + 2))

    return df

data_frame()
```
